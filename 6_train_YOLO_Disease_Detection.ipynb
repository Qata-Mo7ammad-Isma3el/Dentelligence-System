```python
import os
def label_preprocessing(path):
  # Path to the folder containing the .txt files
  folder_path = path

  # Loop through all .txt files in the folder
  for filename in os.listdir(folder_path):
      if filename.endswith(".txt"):
          file_path = os.path.join(folder_path, filename)

          # Read the content of the file
          with open(file_path, "r") as file:
              content = file.read()

          # Replace '/n' with actual newline character '\n'
          cleaned_content = content.replace("/n", "\n")

          # Write the cleaned content back to the same file
          with open(file_path, "w") as file:
              file.write(cleaned_content)

          print(f"Cleaned: {filename}")

  print("✅ All text files processed successfully.")
```

```python
# --- CONFIG: EDIT THESE ---
weights     = "/content/drive/MyDrive/runs_Disease_all/detect/train/weights/best.pt"            # your trained YOLOv8 weights
images_dir  = "/content/drive/MyDrive/dentex_dataset/yolo/disease_all/images/val2017"         # folder with images
labels_dir  = "/content/drive/MyDrive/dentex_dataset/yolo/disease_all/labels/val2017"         # folder with YOLO txt labels (same stems as images)
image_name  = '/content/drive/MyDrive/dentex_dataset/yolo/disease_all/images/val2017/train_249.png'  # e.g. "train_435.png"; or leave None to pick a random image that has a label

# --- Code starts here ---
from pathlib import Path
import random, glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from PIL import Image
from ultralytics import YOLO

def yolo_txt_to_xyxy(label_path, W, H):
    """Returns boxes (N,4) in pixel xyxy and class ids (N,) from a YOLO txt file."""
    if not Path(label_path).exists():
        return np.zeros((0,4), float), np.array([], int)
    boxes, clss = [], []
    with open(label_path, "r") as f:
        for line in f:
            if not line.strip():
                continue
            parts = line.strip().split()
            # robust to prediction txts that include conf: keep first 5 columns
            cls_id, x, y, w, h = map(float, parts[:5])
            x1 = (x - w/2) * W
            y1 = (y - h/2) * H
            x2 = (x + w/2) * W
            y2 = (y + h/2) * H
            boxes.append([x1, y1, x2, y2])
            clss.append(int(cls_id))
    return np.array(boxes, float), np.array(clss, int)

def draw_boxes(ax, boxes, labels=None, scores=None, color="yellow", lw=2):
    for i, (x1, y1, x2, y2) in enumerate(boxes):
        ax.add_patch(Rectangle((x1, y1), x2-x1, y2-y1, fill=False, edgecolor=color, linewidth=lw))
        txt = None
        if labels is not None and scores is not None:
            txt = f"{labels[i]} {scores[i]:.2f}"
        elif labels is not None:
            txt = str(labels[i])
        elif scores is not None:
            txt = f"{scores[i]:.2f}"
        if txt:
            ax.text(x1, max(0, y1-3), txt, fontsize=9, color=color,
                    bbox=dict(facecolor="black", alpha=0.5, pad=1))

# 1) Choose the image
images_dir = Path(images_dir); labels_dir = Path(labels_dir)
if image_name is None:
    # pick a random image that has a corresponding txt
    candidates = []
    for p in images_dir.glob("*"):
        if p.suffix.lower() in {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}:
            if (labels_dir / (p.stem + ".txt")).exists():
                candidates.append(p)
    if not candidates:
        raise FileNotFoundError("No images with matching label txt files were found.")
    img_path = random.choice(candidates)
else:
    img_path = images_dir / image_name
    if not img_path.exists():
        raise FileNotFoundError(img_path)
label_path = labels_dir / (img_path.stem + ".txt")

# 2) Load image & GT labels
img = Image.open(img_path).convert("RGB")
W, H = img.size
gt_boxes, gt_cls = yolo_txt_to_xyxy(label_path, W, H)

# Optional: class names (if you want text labels). Use model.names or define your own list.
model = YOLO(weights)
names = getattr(model, "names", None)  # dict: id->name
gt_labels = [names[c] if (names and c in names) else str(c) for c in gt_cls]

# 3) Predictions
res = model(str(img_path), verbose=False)[0]
pred_xyxy  = res.boxes.xyxy.cpu().numpy() if res.boxes is not None else np.zeros((0,4))
pred_cls   = res.boxes.cls.cpu().numpy().astype(int)    if res.boxes is not None else np.zeros((0,), int)
pred_conf  = res.boxes.conf.cpu().numpy()               if res.boxes is not None else np.zeros((0,))

# (optional) confidence threshold
conf_thres = 0.25
keep = pred_conf >= conf_thres
pred_xyxy, pred_cls, pred_conf = pred_xyxy[keep], pred_cls[keep], pred_conf[keep]
pred_labels = [names[c] if (names and c in names) else str(c) for c in pred_cls]

# 4) Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=120)
axes[0].imshow(img); axes[0].set_title(f"GT — {img_path.name}"); axes[0].axis("off")
axes[1].imshow(img); axes[1].set_title("Prediction"); axes[1].axis("off")

draw_boxes(axes[0], gt_boxes, labels=gt_labels, color="green", lw=2)
draw_boxes(axes[1], pred_xyxy, labels=pred_labels, scores=pred_conf, color="red", lw=2)

plt.tight_layout(); plt.show()
plt.savefig(fname='./')
```

```python
# === CONFIG (edit these) ======================================================
YOLO_WEIGHTS = "/content/drive/MyDrive/runs_Disease_all/detect/train/weights/best.pt"
DATA_YAML    = "/content/drive/MyDrive/configs/yolo/dentex_disease_all_dataset.yaml"
VAL_IMAGES   = "/content/drive/MyDrive/dentex_dataset/yolo/disease/images/val2017"

DINO_CFG     = "/content/drive/MyDrive/configs/dino/DINO_4scale_swin_cls4.py"
DINO_CKPT    = "/content/drive/MyDrive/output_swin_disease_all/checkpoint0027.pth"
save_dir = './'

WBF_IOU_THR  = 0.60      # IoU threshold for fusion
WBF_WEIGHTS  = [2.0, 1.0]# [weight for DINO, weight for YOLO]
CONF_THR     = 0.5      # low conf to let WBF fuse
# =============================================================================

import os, json, glob, math, cv2, numpy as np, torch
from PIL import Image
from ultralytics import YOLO
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import argparse
import torch.serialization
torch.serialization.add_safe_globals([argparse.Namespace])
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Try project WBF first, fallback to ensemble-boxes if needed
try:
    from util.weighted_boxes_fusion import weighted_boxes_fusion
except Exception:
    from ensemble_boxes import weighted_boxes_fusion

# ---- DINO predictor (mirrors your repo) --------------------------------------
import models.dino.datasets.transforms as DT
from util.slconfig import SLConfig

def build_model_main(args):
    from models.registry import MODULE_BUILD_FUNCS
    build_func = MODULE_BUILD_FUNCS.get(args.modelname)
    return build_func(args)

class DinoDetectionPredictor:
    def __init__(self, cfg_path, ckpt_path, score_thr=0.01, use_cuda=True):
        args = SLConfig.fromfile(cfg_path)
        args.device = "cuda" if use_cuda else "cpu"
        model, criterion, post = build_model_main(args)
        ckpt = torch.load(ckpt_path, map_location="cpu")
        model.load_state_dict(ckpt["model"])
        if use_cuda: model = model.cuda()
        model.eval()
        self.model, self.post, self.cuda = model, post, use_cuda
        self.tf = DT.Compose([
            DT.RandomResize([800], max_size=1333),
            DT.ToTensor(),
            DT.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
        ])
        self.score_thr = score_thr

    @torch.no_grad()
    def predict(self, bgr):
        h, w = bgr.shape[:2]
        img = Image.fromarray(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))
        x, _ = self.tf(img, None)
        if self.cuda: x = x.cuda()
        out = self.model(x.unsqueeze(0))
        scale = torch.tensor([(h, w)]).cuda() if self.cuda else torch.tensor([(h, w)])
        o = self.post["bbox"](out, scale)[0]
        m = o["scores"] > self.score_thr
        return (o["boxes"][m].cpu().numpy(),
                o["scores"][m].cpu().numpy(),
                o["labels"][m].cpu().numpy().astype(int))

# ---- STEP A: run YOLO val to get matching COCO GT ----------------------------
GT_JSON = "/content/drive/MyDrive/dentex_dataset/coco/disease/annotations/instances_val2017.json"

yolo = YOLO(YOLO_WEIGHTS)
coco_gt = COCO(GT_JSON)
cat_name_to_id = {c["name"]: c["id"] for c in coco_gt.dataset["categories"]}
model_names = yolo.names

# ---- STEP B: set up predictors ----------------------------------------------
dino = DinoDetectionPredictor(DINO_CFG, DINO_CKPT, score_thr=CONF_THR, use_cuda=torch.cuda.is_available())

def norm_xyxy(boxes, W, H):
    if len(boxes)==0: return np.zeros((0,4), dtype=float)
    b = boxes.astype(float).copy()
    b[:,[0,2]] /= float(W)
    b[:,[1,3]] /= float(H)
    # ensure valid range
    b = np.clip(b, 0.0, 1.0)
    return b

def xyxy_to_xywh(box):
    x1,y1,x2,y2 = box
    return [float(x1), float(y1), float(max(0.0, x2-x1)), float(max(0.0, y2-y1))]

# ---- STEP C: fuse per image and write COCO predictions ----------------------
preds = []
images = coco_gt.dataset["images"]
print(f"Fusing on {len(images)} val images...")

for img_info in images:
    file_name = img_info["file_name"]
    image_id  = img_info["id"]

    img_path = os.path.join(VAL_IMAGES, file_name)
    if not os.path.exists(img_path):
        raise FileNotFoundError(f"Missing image: {img_path}")

    bgr = cv2.imread(img_path)
    H, W = bgr.shape[:2]

    # DINO
    d_boxes, d_scores, d_labels = dino.predict(bgr)          # xyxy, float, int labels

    # YOLO
    y = yolo.predict(bgr, conf=CONF_THR, verbose=False)[0].boxes
    y_boxes  = y.xyxy.cpu().numpy() if len(y) else np.zeros((0,4))
    y_scores = y.conf.cpu().numpy() if len(y) else np.zeros((0,))
    y_labels = y.cls.cpu().numpy().astype(int) if len(y) else np.zeros((0,), dtype=int)

    # Normalize to [0,1] for WBF
    db = norm_xyxy(d_boxes, W, H)
    yb = norm_xyxy(y_boxes, W, H)

    # Labels as python lists
    d_labels_list = d_labels.tolist() if len(d_labels) else []
    y_labels_list = y_labels.tolist() if len(y_labels) else []

    # WBF
    f_boxes, f_scores, f_labels = weighted_boxes_fusion(
        [db, yb],
        [d_scores.tolist(), y_scores.tolist()],
        [d_labels_list, y_labels_list],
        weights=WBF_WEIGHTS, iou_thr=WBF_IOU_THR
    )

    # Denormalize back to pixels
    f_boxes = np.array(f_boxes)
    if f_boxes.size:
        f_boxes[:,[0,2]] *= W
        f_boxes[:,[1,3]] *= H

    # Append in COCO format
    for box, score, cls_idx in zip(f_boxes, f_scores, f_labels):
        cls_idx = int(cls_idx)
        # Map model class name -> COCO category_id to be 100% consistent
        cls_name = model_names[cls_idx]
        cat_id = cat_name_to_id.get(cls_name, None)
        if cat_id is None:
            raise KeyError(f"Class '{cls_name}' not found in GT categories: {list(cat_name_to_id.keys())}")

        xywh = xyxy_to_xywh(box)
        # clip box inside image
        xywh[0] = max(0.0, min(xywh[0], W-1))
        xywh[1] = max(0.0, min(xywh[1], H-1))
        xywh[2] = max(0.0, min(xywh[2], W-xywh[0]-1e-6))
        xywh[3] = max(0.0, min(xywh[3], H-xywh[1]-1e-6))

        preds.append({
            "image_id": int(image_id),
            "category_id": int(cat_id),
            "bbox": [round(v, 3) for v in xywh],
            "score": float(score)
        })

# Save fused predictions
fused_json = os.path.join(save_dir, "fused_predictions.json")
with open(fused_json, "w") as f:
    json.dump(preds, f)
print(f"Saved fused predictions -> {fused_json}")

# ---- STEP D: COCOeval (AP, AP50, AP75, AR@1/10/100) -------------------------
coco_dt = coco_gt.loadRes(fused_json)
e = COCOeval(coco_gt, coco_dt, iouType="bbox")
e.evaluate(); e.accumulate(); e.summarize()

AP       = e.stats[0]  # AP@[.50:.95]
AP50     = e.stats[1]
AP75     = e.stats[2]
AR1      = e.stats[6]  # AR@1
AR10     = e.stats[7]  # AR@10
AR100    = e.stats[8]  # AR@100

print("\n==== FUSED REPORT ====")
print(f"AP@[.50:.95]: {AP:.3f}")
print(f"AP50:        {AP50:.3f}")
print(f"AP75:        {AP75:.3f}")
print(f"AR@1:        {AR1:.3f}")
print(f"AR@10:       {AR10:.3f}")
print(f"AR@100:      {AR100:.3f}")

```

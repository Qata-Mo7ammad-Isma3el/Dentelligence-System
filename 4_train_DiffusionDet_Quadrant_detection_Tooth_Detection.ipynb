# **Evaluate/Test the Model**

```python
import os
import subprocess
import re

# --- SETTINGS ---
config_file = "/content/drive/MyDrive/configs/diffdet/diffdet.dentex.swinbase.quadrant.yaml"
output_dir = "/content/drive/MyDrive/output_diffdet_quadrant"
train_script = "/content/drive/MyDrive/train_diffdet.py"

# --- Collect all matching checkpoints ---
checkpoints = [
    ckpt for ckpt in os.listdir(output_dir)
    if ckpt.endswith(".pth") and (ckpt.startswith("model_") or ckpt == "model_final.pth")
]

# Sort: model_0001234.pth → int(1234), model_final.pth → inf
def extract_iter(ckpt):
    match = re.search(r"model_(\d+)\.pth", ckpt)
    return int(match.group(1)) if match else float('inf')

checkpoints.sort(key=extract_iter)

# --- Evaluate ---
results = {}
for ckpt in checkpoints:
    ckpt_path = os.path.join(output_dir, ckpt)
    print(f"\n🔍 Evaluating {ckpt_path}...")

    cmd = [
        "python", train_script,
        "--config-file", config_file,
        "--eval-only",
        "MODEL.WEIGHTS", ckpt_path
    ]

    try:
        output = subprocess.check_output(cmd, stderr=subprocess.STDOUT, text=True)

        # Parse output line-by-line to find AP table
        lines = output.splitlines()
        found = False
        for i, line in enumerate(lines):
            if "Evaluation results for bbox" in line:
                for j in range(i+1, len(lines)):
                    if "AP50" in lines[j] and "|" in lines[j + 1]:
                        value_line = lines[j + 2] if j + 2 < len(lines) else ""
                        try:
                            # Parse AP, AP50, AP75
                            parts = value_line.strip().split("|")[1:4]
                            ap, ap50, ap75 = map(lambda x: float(x.strip()), parts)
                            results[ckpt] = {"AP": ap, "AP50": ap50, "AP75": ap75}
                            found = True
                            break
                        except Exception as parse_error:
                            continue
                break

        if not found:
            results[ckpt] = "Failed to parse"

    except subprocess.CalledProcessError as e:
        print(f"❌ Error evaluating {ckpt}:\n{e.output}")
        results[ckpt] = "Error"

# --- Print Summary ---
print("\n📊 === Evaluation Summary ===")
for ckpt, val in results.items():
    if isinstance(val, dict):
        print(f"{ckpt}: AP={val['AP']:.2f}, AP50={val['AP50']:.2f}, AP75={val['AP75']:.2f}")
    else:
        print(f"{ckpt}: {val}")

# --- Select Best Model ---
valid_results = {k: v for k, v in results.items() if isinstance(v, dict)}
if valid_results:
    best_ckpt = max(valid_results.items(), key=lambda x: x[1]["AP"])
    print(f"\n🏆 Best Checkpoint: {best_ckpt[0]} — AP={best_ckpt[1]['AP']:.2f}")
else:
    print("\n⚠️ No valid results to select best checkpoint.")
```

```python
import os
import cv2
import uuid
import numpy as np
import matplotlib.pyplot as plt
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data.datasets import load_coco_json
from detectron2.data import MetadataCatalog
from detectron2.engine import DefaultPredictor
from models.diffusiondet.config import add_diffusiondet_config
from models.diffusiondet.util.model_ema import add_model_ema_configs

# --- Remap class index to final label ---
quadrant_remap = {0: 1, 1: 0, 2: 2, 3: 3}
label_map = {0: "1", 1: "2", 2: "3", 3: "4"}

# --- Predictor wrapper ---
class D2DetectionPredictor:
    def __init__(self, cfg, threshold=0.5):
        self.predictor = DefaultPredictor(cfg)
        self.threshold = threshold

    def predict(self, image):
        instances = self.predictor(image)["instances"]
        return instances[instances.scores > self.threshold]

# --- Paths ---
config_file = "/content/drive/MyDrive/configs/diffdet/diffdet.dentex.swinbase.quadrant.yaml"
model_weights = "/content/drive/MyDrive/output_diffdet_quadrant/model_0039999.pth"
image_path = "/content/drive/MyDrive/dentex_dataset/coco/quadrant/val2017/train_249.png"
json_anno = "/content/drive/MyDrive/dentex_dataset/coco/quadrant/annotations/instances_val2017.json"
image_root = "/content/drive/MyDrive/dentex_dataset/coco/quadrant/val2017"
confidence_threshold = 0.01

# --- Load and register dataset ---
dataset_name = f"quadrant_val_{uuid.uuid4().hex[:6]}"
dataset_dicts = load_coco_json(json_anno, image_root, dataset_name)
metadata = MetadataCatalog.get(dataset_name)

# --- Match image to annotation ---
filename = os.path.basename(image_path)
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(image_path)
record = next((d for d in dataset_dicts if os.path.basename(d["file_name"]) == filename), None)
if record is None:
    raise ValueError(f"No ground truth for {filename}")

# --- Load model ---
cfg = get_cfg()
add_diffusiondet_config(cfg)
add_model_ema_configs(cfg)
cfg.merge_from_file(config_file)
cfg.MODEL.WEIGHTS = model_weights
cfg.MODEL.DEVICE = "cuda"
cfg.MODEL.RETINANET.SCORE_THRESH_TEST = confidence_threshold
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = confidence_threshold
cfg.freeze()

predictor = D2DetectionPredictor(cfg, threshold=confidence_threshold)

# --- Inference ---
outputs = predictor.predict(image)

# --- Remap and label ---
pred_classes = outputs.pred_classes.cpu().numpy()
remapped_classes = np.array([quadrant_remap.get(c, c) for c in pred_classes])
labels = [label_map[c] for c in remapped_classes]
outputs.pred_classes = outputs.pred_classes.new_tensor(remapped_classes)

# --- Filter invalid boxes ---
boxes = outputs.pred_boxes.tensor.cpu().numpy()
scores = outputs.scores.cpu().numpy()
keep = []
for i, (box, score) in enumerate(zip(boxes, scores)):
    x0, y0, x1, y1 = box
    w, h = x1 - x0, y1 - y0
    area = w * h
    aspect_ratio = w / h if h > 0 else 0
    if score > 0.3 and 500 < area < 100000 and 0.3 < aspect_ratio < 3.5:
        keep.append(i)

if keep:
    outputs = outputs[keep]

# --- Keep only top 4 by score ---
if len(outputs) > 4:
    top4 = outputs.scores.argsort(descending=True)[:4]
    outputs = outputs[top4]
    labels = [labels[i] for i in top4]

# --- Show visualizations ---
v_gt = Visualizer(image[:, :, ::-1], metadata=metadata, scale=1.0)
gt_img = v_gt.draw_dataset_dict(record).get_image()

# Draw manually with class name labels
v_pred = Visualizer(image[:, :, ::-1], metadata=metadata, scale=1.0)
output_img = v_pred.draw_instance_predictions(outputs.to("cpu")).get_image()

# Add readable quadrant labels manually
for i, box in enumerate(outputs.pred_boxes.tensor.cpu().numpy()):
    x0, y0 = int(box[0]), int(box[1])
    label = labels[i]
    cv2.putText(output_img, label, (x0, y0 - 10), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 0, 0), 5)

pred_img = output_img


# --- Combine and show ---
combined = np.hstack((image[:, :, ::-1], gt_img, pred_img))
plt.figure(figsize=(24, 12))
plt.imshow(combined)
plt.title("Original | Ground Truth | Prediction", fontsize=24, fontweight="bold")
plt.axis("off")
plt.tight_layout()
plt.show()
```
